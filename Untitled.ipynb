```python
orders = pd.read_csv('olist_orders_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')

```

```python
o_dat = orders[['order_id','order_status','order_purchase_timestamp']]
pr_dat = products[['product_id','product_category_name']]
pa_dat = payments[['order_id','payment_value']]
```

```python
o_dat.groupby(['order_status']).count()
```

```python
oi = pd.merge(o_dat, items, how="left", on=["order_id"])
oip = pd.merge(oi, pr_dat, how="left", on=["product_id"])
oipp = pd.merge(oip, pa_dat, how="left", on =['order_id'])
```

```python
oipp.head()
```

```python
oipp['total_payment'] = oipp['payment_value'].groupby(oipp['product_category_name']).transform('sum')
```

```python
f = oipp[['product_category_name','total_payment']]
f = f.drop_duplicates()
f.sort_values(by=['total_payment'], ascending=False)
```
