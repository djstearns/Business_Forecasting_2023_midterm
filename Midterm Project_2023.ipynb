```python
import pandas as pd
```

```python
products = pd.read_csv("olist_products_dataset.csv")
products.head()
```

```python
products.isnull().sum()
```

```python
#Product category name is filled with a other, because it a string, while the others are integers.
products['product_category_name'] = products['product_category_name'].fillna('outro/others')
products = products.fillna(products.median())
```

```python
order_items = pd.read_csv("olist_order_items_dataset.csv")
order_items.head()
```

```python
translation = pd.read_csv("product_category_name_translation.csv")
translation.head()
```

```python
items = pd.read_csv('olist_order_items_dataset.csv')
items.head()
```

```python
payments = pd.read_csv("olist_order_payments_dataset.csv")
payments.head()
```

```python
olist_review = pd.read_csv("olist_order_reviews_dataset.csv")
olist_review.head()
```

```python
#Get missing value number.Missing value are high for comment title and message.
olist_review.isnull().sum()
```

```python
olist_review["review_comment_title"] = olist_review["review_comment_title"].fillna("None")
olist_review["review_comment_message"] = olist_review["review_comment_message"].fillna("No message")
olist_review.head()
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
orders.head()
```

```python
orders.isnull().sum()
```

```python
#forward fill missing values 
orders = orders.fillna(method = 'ffill')
orders.head()
```

```python
time = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for i in time:
    orders[i] = pd.to_datetime(orders[i])

orders.head()
```

```python
sellers = pd.read_csv("olist_sellers_dataset.csv")
sellers.isnull().sum()
```

```python
#Merge Data
dat_a = pd.merge(orders, items, left_on = 'order_id', right_on = 'order_id')
dat_b = pd.merge(dat_a, sellers, left_on = 'seller_id', right_on = 'seller_id')
dat_c = pd.merge(dat_b, products, left_on = 'product_id', right_on = 'product_id')
dat_d = pd.merge(dat_c, olist_review, left_on = 'order_id', right_on = 'order_id')
dat_e = pd.merge(dat_d, translation, left_on = 'product_category_name', right_on = 'product_category_name')
dat_frame = pd.merge(dat_e, payments, how = 'left', left_on = 'order_id', right_on = 'order_id')

dat_frame.head()
```

We have 2 columms, order_purchase_timestamp,price.

```python
#sort dat_frame, using order_purchase_timestamp,extract the date from date timestamp,
#reset the index and drop unwanted columns.
monthfcast_3 = dat_frame[['order_purchase_timestamp','price']]
monthfcast_3.head()
```
